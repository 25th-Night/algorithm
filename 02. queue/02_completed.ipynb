# Array

```python
class Queue():
    def __init__(self):
        self.queue = []
    
    def isEmpty(self):
        is_empty = False
        if len(self.queue) == 0:
            is_empty = True
        return is_empty

    def enqueue(self, data):
        self.queue.append(data)

    def dequeue(self):
        deque_object = None
        if self.isEmpty():
            print("Queue is empty")
        else:
            deque_object = self.queue[0]
            self.queue = self.queue[1:]
        return deque_object

    def peek(self):
        peek_object = None
        if self.isEmpty():
            print("Queue is empty")
        else:
            peek_object = self.queue[0]
        return peek_object
```

## 스택과의 차이를 이해해야 함

> `printQueue` 구현 시 `printStack` 과의 차이점 생각하기

- 큐를 연결 리스트로 구현할 때, 앞(front)에서 제거하고, 뒤(back)에서 삽입하는 방식으로 동작하며, front와 back을 사용하여 큐의 시작과 끝을 추적함

### 큐의 정의 및 코드 구현

- 1 ← 2 ← 3

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedListQueue:
    def __init__(self):
        self.front = None
        self.back = None

    def isEmpty(self):
        is_empty = False
        if self.front is None:
            is_empty = True
        return is_empty
    
    def printQueue(self):
        current_node = self.head
        string = ""

        while current_node:
            string += str(current_node.data)
            current_node = current_node.next
            if current_node:
                string += " -> "
        print(string)
    
    def enqueue(self, data):
        new_node = Node(data)
        if self.front is None:
            self.front = new_node
            self.back = new_node
        else:
            self.back.next = new_node
            self.back = self.back.next

    def dequeue(self):
        deque_object = None
        if self.isEmpty():
            print("Queue is empty")
        else:
            deque_object = self.front.data
            self.front = self.front.next
        return deque_object

    def peek(self):
        peek_object = None
        if self.isEmpty():
            print("Queue is empty")
        else:
            peek_object = self.front.data
        return peek_object
```
